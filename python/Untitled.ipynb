```python
cl_params = ['32','64','128','256','512'] # 32 is special
```

```python
log32 = open('q32withbase.log')
datapoints = ["acc: 0.574775, f1: 0.363112, aupr: 0.481019"]
for line in log32.readlines():
    if line[:3] == 'acc':
        datapoints.append(line)
log32.close()
```

```python
baseline = datapoints[:5]
points32 = datapoints[5:]
```

```python
result = ""
for point in baseline:
    [_, acc, _, f1, _, aupr] = point.split()
    result += f'{acc[:-1]}\t{f1[:-1]}\t{aupr[:-1]}\n'
print(result)
```

```python
def print_result(points):
    result = ""
    for point in points:
        [_, acc, _, f1, _, aupr] = point.split()
        result += f'{acc[:-1]}\t{f1[:-1]}\t{aupr[:-1]}\n'
    print(result)
```

```python
print_result(points32[25:])
```

```python
code = ""
accs = []
f1s = []
auprs = []
for point in baseline:
    [_, acc, _, f1, _, aupr] = point.split()
    accs.append(float(acc[:-1]))
    f1s.append(float(f1[:-1]))
    auprs.append(float(aupr[:-1]))
code += f'base.acc = {accs};\n'
code += f'base.f1 = {f1s};\n'
code += f'base.aupr = {auprs};\n'

for cl in cl_params:
    log = open(f'q{cl}.log')
    datapoints = []
    for line in log.readlines():
        if line[:3] == 'acc':
            datapoints.append(line)
    log.close()
    
    for i in range(int(len(datapoints) / 5)):
        points = datapoints[i*5:(i+1)*5]
        accs = []
        f1s = []
        auprs = []
        for point in points:
            [_, acc, _, f1, _, aupr] = point.split()
            accs.append(float(acc[:-1]))
            f1s.append(float(f1[:-1]))
            auprs.append(float(aupr[:-1]))
        code += f'pen{cl}.cl{2**(i+1)}.acc = {accs};\n'
        code += f'pen{cl}.cl{2**(i+1)}.f1 = {f1s};\n'
        code += f'pen{cl}.cl{2**(i+1)}.aupr = {auprs};\n'
```

```python
print(code)
```
